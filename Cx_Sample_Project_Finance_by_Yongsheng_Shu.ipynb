## Q2: Top 3 highest percentage gain ticker for 5-year investment in 2013-2017

```python
gain_5_years = {}

for ticker in ticker_list:

    df_1 = df[df.Name == ticker]
    #print(df_1)
    
    for d in df_1.date:
        if d >= "2013-01-01":
            d0 = d
            break
    for d in df_1.date:
        if d >= "2018-01-01":
            d3 = d
            break

    
    df_2 = df_1[df_1.date==d0]
    #print(df_2)   
    p0 = (list(df_2.open))[0]
    #print(p0)
    
    df_2 = df_1[df_1.date==d3]
    #print(df_2)
    p3 = (list(df_2.open))[0]
    #print(p3)

    g = round((p3 - p0)/p0*100, 2)

    gain_5_years[ticker] = g
    
    #break
    
#print(gain_2015, gain_2016, gain_2017)
print(len(gain_5_years))

#print(gain_5_years)
```

```python
import operator
gain_5_years_sort = sorted(gain_5_years.items(), key=operator.itemgetter(1), reverse=True)

print(type(gain_5_years_sort)) # gain_5_years_sort is a list of list, not a dictionary
#print(gain_5_years_sort)
```

## Q4: Highest percentage gain tickers for 2015, 2016, 2017 of the top 10 trading volume tickers in 2014

### Top 10 trading volume tickers in 2014

```python
volume_dict = {}

for ticker in ticker_list:

    df_1 = df[df.Name == ticker]
    #print(df_1)
    
    total = 0
    i = 1
    
    for d in df_1.date:
        if d >= "2014-01-01" and d <= "2014-12-31":
            total += list(df_1["volume"])[0]
            i += 1
            
    #print(ticker, i, total/i)
    
    volume_dict[ticker] = int(total/i)
    
    #break
    
#print(volume_dict)

import operator
volume_dict_sort = sorted(volume_dict.items(), key=operator.itemgetter(1), reverse=True)

print(type(volume_dict_sort)) # volume_dict_sort is a list of list, not a dictionary
```

```python
print(volume_dict_sort[0][0])
print(volume_dict_sort[0][1])

top_10_volume_list_2014 = [volume_dict_sort[i][0] for i in range(10)]
print(top_10_volume_list_2014)
```
