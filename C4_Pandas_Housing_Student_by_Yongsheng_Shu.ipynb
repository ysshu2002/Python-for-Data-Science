```python
"""
Data from https://data.ftb.ca.gov/California-Personal-Income-Tax/Personal-Income-Tax-Statistics-By-Zip-Code/mriu-wsxf
The average house income by ZipCode can be calculated:
94085: 100007
94087: 164080
95014: 206328
95051: 130611
"""
```

```python
"""
Assuming that the house incomes of every home in a ZipCode are the same, which is the above average house income by ZipCode.
Then a new column "House_Income" can be added in the Housing Dataframe

And assuming that the house incomes of every home not in the above 4 ZipCode is 150000

"""
```

```python
import numpy as np
df['House_Income'] = np.where(df['ZipCode'] == 94085,
                              100007, np.where(df['ZipCode'] == 94087,
                                               164080, np.where(df['ZipCode'] == 95014,
                                                                206328, np.where(df['ZipCode'] == 95051,
                                                                                 130611, 150000))))
```

```python
df.head()
```

```python
"""
Now let's see Price per House Income by ZipCode
"""
```

```python
df["Price_per_House_Income"] = df.HomePriceK / df.House_Income
df.groupby("ZipCode").Price_per_House_Income.describe()
```
